```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Load the train and test datasets
train_data_path = '/content/drive/MyDrive/train dataset/train_data.csv'  # Replace with the actual path to train_data.csv
test_data_path = '/content/drive/MyDrive/test dataset/test_data.csv'    # Replace with the actual path to test_data.csv

train_df = pd.read_csv(train_data_path)
test_df = pd.read_csv(test_data_path)

# Preprocess the data (if needed)
# Handle missing values, encode categorical variables, etc.

# Split the data into features and target variable
X_train = train_df.drop(columns=['Survived'])
y_train = train_df['Survived']

X_test = test_df.drop(columns=['Survived'])
y_test = test_df['Survived']

# Train the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate accuracy and print the confusion matrix
accuracy = model.score(X_test, y_test)
conf_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])

print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(conf_matrix)

# Plot the cost versus the number of iterations graphically
cost_values = []
for i in range(1, 1001):
    model.set_params(max_iter=i)
    model.fit(X_train, y_train)
    cost_values.append(-model.decision_function(X_train).mean())

iterations = np.arange(1, 1001)

plt.plot(iterations, cost_values)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs. Iterations for Logistic Regression')
plt.show()
```
